```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =pd.read_excel('/content/drive/MyDrive/Colab Notebooks/covid.xlsx')
data
```

```python
data.isnull().sum()
```

```python
import plotly.express as px
```

```python
fig = px.bar(data, x='Date_YMD',y='Daily confirmed cases')
fig.show()
```

```python
cases = data["Daily confirmed cases"].sum()
deceased = data["Deceased"].sum()

labels = ["confirmed", "Deceased"]
values = [cases, deceased]

fig = px.pie(data, values=values, 
             names=labels, 
             title='Daily confirmed cases vs Daily Deaths', hole=0.5)
fig.show()
```

```python
death_rate = (data["Deceased"].sum() / data["Daily confirmed cases"].sum()) * 100
print(death_rate)
```
